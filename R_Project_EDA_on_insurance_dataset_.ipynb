# **Adjusting the X and Y axis limits**
Method 1: By deleting the points outside the range

```r
g <- ggplot(df, aes(x=age, y=expenses)) + geom_point() + geom_smooth(method="lm")  # set se=FALSE to turnoff confidence bands

# Delete the points outside the limits
g + xlim(c(18,30)) + ylim(c(0,60000))   # deletes points
```

Method 2: Zooming In

```r
# Zoom in without deleting the points outside the limits.
# As a result, the line of best fit is the same as the original plot.
g1 <- g + coord_cartesian(xlim=c(18,30), ylim=c(0,60000))  # zooms in
plot(g1)
```

How to Change the Title and Axis Labels

```r
g1 + labs(title="age Vs expenses",
subtitle="From insurance dataset",
y="expenses",
x="age",
caption="insurance Demographics")
```

How to Change the Color and Size of Points

```r
ggplot(df, aes(x=age, y=expenses)) + geom_point(col="green", size=2) +   # Set static color and size for points
  geom_smooth(method="lm", col="red",se=FALSE) +  # change the color of line
  coord_cartesian(xlim=c(18,35), ylim=c(0,60000)) +
  labs(title="age Vs expenses", subtitle="From insurance dataset", y="expenses", x="age", caption="insurance Demographics")
```

How to Change the Color To Reflect Categories in Another Column

```r
gg <- ggplot(df, aes(x=age, y=expenses)) +
  geom_point(aes(col=sex), size=2) +  # Set color to vary based on state categories.
  geom_smooth(method="lm", col="firebrick", size=1,se=FALSE) +
  coord_cartesian(xlim=c(18,35), ylim=c(0,60000)) +
  labs(title="age Vs expenses", subtitle="From insurance dataset", y="expenses", x="age", caption="insurance Demographics")
plot(gg)
```
